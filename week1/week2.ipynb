```python
import random

def randomizer(start, end):
  """Generates a random integer between start and end (inclusive).

  Args:
    start: The lower bound of the random number range.
    end: The upper bound of the random number range.

  Returns:
    A random integer between start and end (inclusive).
  """
  return random.randint(start, end)

# Example usage:
random_number = randomizer(1, 100)
print(random_number)
```
